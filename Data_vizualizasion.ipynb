```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use("classic")
def f(x, y):
  return np.sin(x)**10 + np.cos(10+x+y)+ np.cos(x)

x = df["SibSp"]
y = df["Fare"]
x, y = np.meshgrid(x, y)
z = f(x, y)
plt.figure(dpi=150)
plt.contour(x, y, z, 20, cmap="RdGy")
plt.show()
```

```python
import seaborn as sns

with sns.axes_style("darkgrid"):
  plt.figure(dpi=100)
  sns.kdeplot(data=df, x="Survived", y="Pclass")

  plt.figure(dpi=100)
  sns.kdeplot(df.loc[df["Survived"]==0, 'Pclass'], shade=True)
  sns.kdeplot(df.loc[df["Survived"]==1, 'Pclass'], shade=True)
  plt.legend( )
  sns.jointplot(data=df, x="Survived", y="Pclass", kind='hex', height=7)
```
