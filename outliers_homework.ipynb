```python
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_csv("/content/database.csv")
```

```python
df
```

```python
df = df.select_dtypes(include="number")
```

```python
IF = IsolationForest()
```

```python
df["outlier"] = pd.Series(IF.fit_predict(df.values))
```

```python
df = df[df.outlier!=-1]
df = df.drop(["outlier"], axis=1, inplace=True)
```

```python
df
```

```python
from sklearn.covariance import EllipticEnvelope
```

```python
ee = EllipticEnvelope()
```

```python
df["outlier"] = pd.Series(ee.fit_predict(df.values))
df = df[df.outlier!=-1]
df.drop(["outlier"], axis=1, inplace=True)
```

```python
df
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
df = df.select_dtypes(include="number")
lof = LocalOutlierFactor()
df["outlier"] = pd.Series(lof.fit_predict(df.values))
df = df[df.outlier!=-1]
df = df.drop(["outlier"], axis = 1, inplace=True)
```

```python
from sklearn.svm import OneClassSVM
```

```python
svm = OneClassSVM()
df["outlier"] = pd.Series(svm.fit_predict(df.values))
df = df[df.outlier!=-1]
df = df.drop("outlier", axis=1, inplace=True)
```
